```python
valores_ausentes = ['***', '###!', '****', '*****', 'NULL']

df = pd.read_csv('ocorrencia.csv', sep= ';', parse_dates=['ocorrencia_dia'], dayfirst=True, na_values=valores_ausentes)

df
```

```python
flt_incidentes = df.ocorrencia_classificacao == "INCIDENTE"

flt_incidentes_graves = df.ocorrencia_classificacao == "INCIDENTE GRAVE"

flt_acidente = df.ocorrencia_classificacao == "ACIDENTE"
```

```python
flt_acidente = df[(df["ocorrencia_classificacao"] == "ACIDENTE")]
flt_acidente
```

```python
flt_2012 = (df.ocorrencia_dia >="2012-01-01") & (df.ocorrencia_dia <="2012-03-31")
flt_2013 = (df.ocorrencia_dia >="2013-01-01") & (df.ocorrencia_dia <="2013-03-31")
flt_2014 = (df.ocorrencia_dia >="2014-01-01") & (df.ocorrencia_dia <="2014-03-31")
```

```python
flt_2017 = df.ocorrencia_dia.dt.year == 2017
flt_2019 = df.ocorrencia_dia.dt.year == 2019
flt_rio = df.ocorrencia_cidade.str[:3] == "RIO"
flt_uf = df.ocorrencia_uf == 'RJ'
```

```python
df_1 = df.loc[(flt_2012 | flt_2013 | flt_2014) & flt_uf]
rj = df_1[["ocorrencia_classificacao", "ocorrencia_uf"]]
rj.value_counts()
```

```python
df_1.groupby(['ocorrencia_cidade', 'ocorrencia_classificacao']).count()
```

```python
df_1.groupby(['ocorrencia_cidade', 'ocorrencia_classificacao']).size()
```

```python
df_1.groupby(['ocorrencia_cidade', 'ocorrencia_classificacao']).size().sort_values(ascending=False)
```

```python
df_1.sort_values(by="ocorrencia_cidade")
```

contatenação de DFs

```python
f_1 = df.ocorrencia_uf == 'MG'
f_1 = df.loc[f_1, ["ocorrencia_classificacao", "ocorrencia_cidade", "ocorrencia_uf"]]

f_2 = df.ocorrencia_uf == 'SP'
f_2 = df.loc[f_2, ["ocorrencia_classificacao" , "ocorrencia_cidade"]]


df_MG_SP = pd.concat([f_1, f_2])
df_MG_SP
```
