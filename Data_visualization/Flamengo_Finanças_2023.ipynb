```python
import matplotlib.pyplot as plt

# Dados
times = ["Flamengo", "Manchester City", "Real Madrid", "Bayern de Munique", "Liverpool"]
receitas = [1374, 731, 640, 611, 594]

# Cores
flamengo_cor = "#9C0000"
outros_clubes_cor = "#0000CD"

# Gráfico de barras
plt.bar(times, receitas, color=flamengo_cor, label="Flamengo")  # Add label for Flamengo bar
plt.bar(times[1:], receitas[1:], color=outros_clubes_cor, label="Outros Clubes")  # Add label for other clubs

# Personalização
plt.title("Comparação entre Flamengo e Money League (2023)")
plt.xlabel("Times")
plt.ylabel("Receitas (€ milhões)")
plt.xticks(rotation=90)

# Legend (even if only one labeled element)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))  # Explicit legend call

# Adicionar valores específicos
for i, receita in enumerate(receitas):
    plt.annotate(f"{receita}", (i, receita), ha="center", va="bottom")

# Adicionar linha de referência para média
media_receitas = sum(receitas) / len(receitas)
plt.axhline(y=media_receitas, color="gray", linestyle="dashed")
# plt.annotate("Média Money League", (1,media_receitas))
# Add a text annotation with the average revenue value
plt.annotate("Média Money League: {:.2f}".format(media_receitas), (3, media_receitas),ha="right")


# Exibir o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt

# Dados
anos = ["2019", "2020/A", "2021/A", "2022", "2023"]
caixa = [112, 80, 169, 266, 263]
endividamento = [466, 585, 297, 227, 48]

# Cores
caixa_cor = "black"
endividamento_cor = "#FF0000"

# Criar lista com os valores percentuais em relação a 2019
percentuais = [round(((valor - endividamento[0]) / endividamento[0]),2) * 100 for valor in endividamento[1::]]

print(percentuais)


# Gráfico combinado
plt.plot(anos, caixa, color=caixa_cor, label="Caixa")
plt.bar(anos, endividamento, bottom=0, color=endividamento_cor, label="Endividamento")

for i, percentual in enumerate(percentuais):
    plt.annotate(f"{percentual:.2f}%", (i, endividamento[i]), ha="center", va="bottom")



# Personalização
plt.title("Relação entre Caixa e Endividamento do Flamengo")
plt.xlabel("Ano")
plt.ylabel("Valor (R$ milhões)")
plt.xticks(anos)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

# Adicionar linhas de referência (opcional)
# plt.axhline(y=media_caixa, color="gray", linestyle="dashed")
# plt.axhline(y=media_endividamento, color="gray", linestyle="dashed")

# Exibir o gráfico
plt.show()
```
