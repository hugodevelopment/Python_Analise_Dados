<h3> Kpis Resultados Flamengo </h3>

* Receita Bruta: **R$ 1.374 bilhão**

* Receita Recorrente: **R$ 1.071 bilhão**

* Ebitda: **R$ 485 milhões**

* Superávit: **R$ 320 milhões**

* Dívida Operacional Líquida: **R$ 48 milhões**

<h3> Analisando Individualmente </h3>

1. Receita Bruta:

Definição: Valor total de todos os rendimentos do Flamengo em 2023, incluindo vendas de produtos, direitos de transmissão, patrocínios, etc.
Indicadando força da marca Flamengo e sua capacidade de gerar renda.

Em 2023: Atingiu R$ 1,374 bilhão, um recorde histórico para o clube.
Comparação: **Aumento de 27% em relação a 2022.**

2. Receita Recorrente:

Definição: Receitas previsíveis e contínuas do Flamengo, excluindo transações pontuais como venda de jogadores.
Indicabdo sustentabilidade financeira do clube a longo prazo.
Em 2023: Atingiu R$ 1,071 bilhão.
Comparação: **Aumento de 25% em relação a 2022.**

3. Ebitda:

Definição: Lucro antes de juros, impostos, depreciação e amortização.
Indicando lucratividade operacional do Flamengo, sem considerar os efeitos de financiamentos e investimentos.
Em 2023: Atingiu R$ 485 milhões.
Comparação: **Aumento de 31% em relação a 2022.**

4. Superávit:

Definição: Diferença positiva entre as receitas e as despesas do Flamengo.
Indicador de: Capacidade do clube de gerar caixa e reinvesti-lo.
Em 2023: Atingiu R$ 320 milhões, o maior da história do Flamengo.
Comparação: **Mais do que o dobro do valor de 2022.**

5. Dívida Operacional Líquida:

Definição: Dívida total do Flamengo, menos caixa e outros ativos líquidos.
Indicador de: Endividamento do clube e sua capacidade de honrar seus compromissos.
Em 2023 Reduziu de 227 milhões para R$ 48 milhões.
Comparação: **Queda de 79% em relação a 2022.**

**Conclusão:**

Os KPIs do Flamengo em 2023 demonstram um desempenho financeiro excepcional, com aumento de receitas, lucratividade e superávit, além de uma significativa redução da dívida. Isso coloca o clube em uma posição de solidez e destaque no cenário futebolístico nacional e internacional.

<h3> Analisando Gráficamente </h3>

<h3> 1º Caixa e Endividamento </h3>

Analisando o resultado demostrativo do Flamengo temos o seguinte gráfico:

1.png

Podemos perceber que há uma relação entre o saldo em caixa e a redução do endividamento. Pensando nisso fiz um gráfico relacionando essas duas grandezas e calculando o percentual referente a 2019.  

Percebemos assim que em relação a 2019 o endividamento teve uma redução de 90% do valor.




```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
anos = ["2019", "2020/A", "2021/A", "2022", "2023"]
caixa = [112, 80, 169, 266, 263]
endividamento = [466, 585, 297, 227, 48]

# Cores
caixa_cor = "black"
endividamento_cor = "#FF0000"

# Criar lista com os valores percentuais em relação a 2019
percentuais = [round(((valor - endividamento[0]) / endividamento[0]),2) * 100 for valor in endividamento]


# Gráfico combinado
plt.plot(anos, caixa, color=caixa_cor, label="Caixa")
plt.bar(anos, endividamento, bottom=0, color=endividamento_cor, label="Endividamento")

for i, percentual in enumerate(percentuais):
    plt.annotate(f"{percentual:.2f}%", (i, endividamento[i]), ha="center", va="bottom")


for i, valor in enumerate(caixa):
    plt.annotate(f"{valor:.1f}", (i, caixa[i]), ha="center", va="bottom")



# Personalização
plt.title("Relação entre Caixa e Endividamento do Flamengo")
plt.xlabel("Ano")
plt.ylabel("Valor (R$ milhões)")
plt.xticks(anos)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))


# Exibir o gráfico
plt.show()
```

<h3> Comparação de  Receitas de Matchday entre Flamengo e Money League 2023 <h3>

**Comparando com as receitas de Matchday dos times da Money League o Flamengo se encontra na 18º com 56% da mediana do valor dos clubes.**






Neste caso a mediana dos valores é 83 milhões de dolares. No gráfico abaixo os times que estão acima da mediana estão em azul, enquanto os times abaixo da mediana estão em vermelho, destacando Flamengo no gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
clubes = ["Barcelona", "Psg", "Tottehan", "Man United", "Real Madrid", "Bayer Mun", "Arsenal", "Liverpool", "Inter de Milão", "Chealsea", "Man City", "Milan", "Borussia", "Olym de Mar", "Atl de Madrid", "Juventus", "Frankfurt", "Flamengo", "Wst Ham", "New Castle","Napoli"]
receitas = [166, 153, 151, 135, 122, 121, 118, 103, 90, 88, 83, 79, 75, 64, 60, 57, 54, 48, 47, 44, 44]

# Cálculo da Mediana
mediana = np.median(receitas)

# Posição da barra do Flamengo
flamengo_posicao = 17

# Lista com as porcentagens
porcentagens = [round(((receita / mediana)) * 100,1) for receita in receitas]

# Cores para as barras
cores = ["blue" if receita >= mediana else "darkred" for receita in receitas]

# Gráfico de Barras
plt.figure(figsize=(15, 8))
plt.bar(clubes, receitas, color=cores, width=0.5)
# Destacando a barra flamengo
plt.bar(clubes[flamengo_posicao], receitas[flamengo_posicao], color="black" if receitas[flamengo_posicao] else cores, width=0.5)

# Adicionar valores específicos
for i, receita in enumerate(receitas):
    plt.annotate(f"{receita} ({porcentagens[i]}%)",
                 (i-0.60, receita),
                 ha="center",
                 va="baseline",
                 fontsize=9,
                 bbox=dict(boxstyle="round,pad=0.1", fc="white"))


# Personalização
plt.title("Comparação entre Flamengo e clubes da Money League – receitas de matchday (2023) (€ milhões)")
plt.xlabel("Clube")
plt.ylabel("Receitas (R$ milhões)")
plt.xticks(clubes, rotation=35)

# Mediana
plt.axhline(y=mediana - 1, color="gray", linestyle="dashed")
plt.annotate("Média Money League: {:.2f}".format(mediana), (15, mediana), ha="left")

# Legenda
plt.legend(loc="best", title="Flamengo representa aproximadamente 57% da mediana dos valores")

# Limite do eixo y
plt.ylim(bottom=0, top=max(receitas) * 1.1)

# Exibir o gráfico
plt.show()
```

<h3> Comparação de Receitas Broadcast entre Flamengo e Money League 2023 <h3>

No gráfico abaixo representa o percentual de cada clube no valor total das receitas, podemos perceber que a menor percentual é do Flamengo, com menos de 2%.

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados
clubes = ["Man City", "Real Madrid", "Liverpool", "Chealse", "Psg", "Man Untd", "Tottenhan", "Barcelona", "Arsenal", "Inter de Milão", "Bayer", "New Castle", "Milan", "Atl Madri", "West Ham", "Napoli", "Borussia", "Juventus","Frankfurt","Olym Mars","Flamengo"]
receitas = [344,306,282,260,249, 240, 235,222, 220, 216,204,190,182, 182,170, 163, 158,157,141,138,79]


# Cálculo da Mediana
soma = np.sum(receitas)

# Posição da barra do Flamengo
flamengo_posicao = 20

# Lista com as porcentagens
porcentagens = [round(((receita / soma)) * 100,1) for receita in receitas]

cores = "darkred"

# Gráfico de Barras
plt.figure(figsize=(15, 8))
plt.bar(clubes, receitas, color= cores, width=0.5)
# Destacando a barra flamengo
plt.bar(clubes[flamengo_posicao], receitas[flamengo_posicao], color= "black" if receitas[flamengo_posicao] else cores, width=0.5)


# Adicionar valores específicos
for i, receita in enumerate(receitas):
    plt.annotate(f"{receita} ({porcentagens[i]}%)",
                 (i, receita),
                 ha="center",
                 va="baseline",
                 fontsize=9,
                 bbox=dict(boxstyle="round,pad=0.1", fc="white"))


# Personalização
plt.title("Percentual de participação entre Flamengo e clubes da Money League nas receitas de comerciais totais (2023) (€ milhões)")
plt.xlabel("Clube")
plt.ylabel("Receitas (R$ milhões)")
plt.xticks(clubes, rotation=35)

# Mediana
plt.axhline(y=soma - 1, color="gray", linestyle="dashed")
plt.annotate("Médiana dos valores Money League: {:.2f}".format(soma), (15, soma), ha="left")

# Legenda
plt.legend(loc="best", title="Flamengo representa aproximadamente 1.6% da soma total dos valores de receita comercial")

# Limite do eixo y
plt.ylim(bottom=0, top=max(receitas) * 1.1)

# Exibir o gráfico
plt.show()
```
