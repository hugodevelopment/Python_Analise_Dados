```python
import pandas as pd
import matplotlib.pyplot as plt

# Criando um dicionário com os nomes dos candidatos e as porcentagens dos votos
data = {"Presidentes": ["Alberto Fernández", "Cristina Kirchner", "Néstor Kirchner", "Macri", "Milei"],
        "total": [4.3, 11.9, 12.9, 13, 14.5]}

# Criando um dataframe a partir do dicionário
df = pd.DataFrame (data)

ordenando = df.sort_values(by="total")

# Criando um gráfico de barras com os dados do dataframe
ax = ordenando.plot.barh (x="Presidentes", y="total", color=["darkorange", "darkred", "red", "darkblue", "blue"])

# Adicionando um título ao gráfico
plt.title("Presidentes e votos")

# Adicionando uma legenda ao gráfico
ax.legend(["Porcentagem de votos%"])

# Adicionando rótulos às barras
for i in ax.patches:
    ax.text(i.get_width()-3., i.get_y()+.25, \
            str(round((i.get_width()), 2)), fontsize=10, color='white')

# Mostrando o gráfico na tela
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Carregue o shapefile da Argentina (você precisa ter um shapefile das províncias da Argentina)
argentina = gpd.read_file('provincias.shp')


# Dicionário com os dados dos candidatos
dados = {
    "Sergio Massa": {
        "cor": "red",
        "votos": {
            "Chaco": 49.9,
            "Buenos Aires": 50.7,
            "Córdoba": 25.9,
            "Mendoza": 28.9,
            "Santa Cruz": 41.9,
            "Santa Fe": 37.2,
            "Ciudad Autónoma de Buenos Aires": 42.8,
            "Chubut":40.9,
            "Corrientes":46.8,
            "Entre Ríos":38.5,
            "Formosa":56.5,
            "Jujuy":41.7,
            "La Pampa":42.6,
            "La Rioja":46.3,
            "Misiones":43.2,
            "Neuquén":30.6,
            "Río Negro": 45.8,
            "Salta:": 42.1,
            "San Juan": 39.3,
            "San Luis": 32,
            "Santiago del Estero":68.4,
            "Terra do Fogo":46.7,
            "Tucumán":48
        }
    },
    "Javier Milei": {
        "cor": "blue",
        "votos": {
            "Chaco": 50.1,
            "Buenos Aires": 49.3,
            "Córdoba": 74.1,
            "Mendoza": 71.1,
            "Santa Cruz": 58.1,
            "Santa Fe": 62.8,
            "Ciudad Autónoma de Buenos Aires": 57.2,
            "Chubut":59.1,
            "Corrientes":53.2,
            "Entre Ríos":61.5,
            "Formosa":43.5,
            "Jujuy":58.3,
            "La Pampa":57.4,
            "La Rioja":53.7,
            "Misiones":56.8,
            "Neuquén":69.4,
            "Río Negro": 54.2,
            "Salta:": 57.9,
            "San Juan": 60.7,
            "San Luis": 68,
            "Santiago del Estero":31.6,
            "Terra do Fogo":53.3,
            "Tucumán":52
        }
    }
}

import numpy as np

for candidato, info in dados.items():
    argentina[candidato] = argentina['NAM'].map(info['votos'])
  

# Crie uma nova coluna para o vencedor em cada província
argentina['Vencedor'] = np.where(argentina['Sergio Massa'] > argentina['Javier Milei'], 'Sergio Massa', 'Javier Milei')

# Crie um dicionário para mapear cada candidato a uma cor
cores = {'Sergio Massa': 'blue', 'Javier Milei': 'red'}

# min_x, min_y, max_x, max_y = argentina.total_bounds


# Crie uma nova figura e eixos
fig, ax = plt.subplots(1, 1)

# Defina os limites do gráfico para focar apenas nas áreas terrestres
ax.set_xlim(-100,-20) # data coordinates
ax.set_ylim(-55,-20)


# Crie o mapa para o vencedor em cada província, usando o dicionário de cores
argentina.plot(column='Vencedor', ax=ax, legend=True, cmap=ListedColormap(cores.values()), categorical=True)


# Defina o título do gráfico
ax.set_title('Vencedor por Província')

# Ajuste o layout para garantir que tudo se encaixe
plt.tight_layout()

# Mostre a figura
plt.show()
```
